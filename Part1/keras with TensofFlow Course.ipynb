```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, BatchNormalization, Conv2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python

# DATA SET GENERATION
train_labels = []
train_samples= []

# total 2100 candidates 

for i in range(50):
    random_younger=randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)# 1 for sideeffect

    random_older= randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)# 0 for not sideeffect


for i in range(1000):
    random_younger=randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)# 1 for sideeffect

    random_older= randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)# 0 for not sideeffect


```

```python
for i in range(0,len(train_labels)):
    print(train_samples[i],train_labels[i])
```

## Preporcessing Data

```python
#convert to np array
train_labels=np.array(train_labels)
train_samples=np.array(train_samples)

# pre shuffle samples
train_labels, train_samples=shuffle(train_labels, train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))


scaled_train_samples
```

# Create NN

```python
model= Sequential([
    Dense(units=16,input_shape=(1,),activation='relu'),
    Dense(units=32,activation='relu'),
    Dense(units=2,activation='softmax'),
])
```

```python
# model.save('./modkl')
model.summary()
```

# Training On Data

```python
model.compile(optimizer=Adam(learning_rate=.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=scaled_train_samples,y=train_labels,batch_size=10,epochs=30,shuffle=True,verbose=2)
```

```python
model.fit(x=scaled_train_samples,y=train_labels,batch_size=10,epochs=30,shuffle=True,verbose=2,validation_split=.1)
```
